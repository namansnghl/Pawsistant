```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext,PromptTemplate
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import SimpleInputPrompt

# VectorStoreIndex = for storing chunks as vectors
# SimpDirc Reader = to read documents
# ServiceContext = bundle of commonly used resources used during the indexing and querying stage in a LlamaIndex pipeline/application
# PrmptTemplate = template for the user prompt to the LLM
# SimpleInputPrompt = style of the prompt used
```

```python
docs = SimpleDirectoryReader('/content/data').load_data()
docs
```

```python
sys_prompt = """
You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided.
"""
# use defaulta format that is supported by llama 2

query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python

import torch

llm = HuggingFaceLLM(
    context_window=4096, # total context window
    max_new_tokens=256, # max token the model can generate along with the context
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    # temperature: how creative the model can be with the output
    # do_sample: model generates text deterministically by always picking the most probable token
    system_prompt=sys_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # if using CUDA to reduce memory usage
    # model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True}
)
```

Once we have the LLM Setup, we move on to sentence embedding of the documents that are fed to the LLM \
## Set up an embedding model by:

- Loading a Hugging Face model (all-mpnet-base-v2) to turn text into numerical vectors.
- Wrapping it in LangchainEmbedding so it works with LlamaIndex.
- The final embed_model can now be used for tasks like searching, clustering, or analyzing text in a structured way.

```python
from langchain.embeddings import HuggingFaceEmbeddings #Lets you use Hugging Face models to turn text into numbers (embeddings).
from llama_index.core import ServiceContext, Settings #A setup in LlamaIndex to manage settings like embeddings and model configurations.
from llama_index.embeddings.langchain import LangchainEmbedding # Connects LangChain text embeddings to work with LlamaIndex.

embed_model=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
from llama_index.core import Settings

# Configure global settings
Settings.llm = llm
Settings.embed_model = embed_model
Settings.chunk_size = 1024 # chunk size for all the chunks made from the documents

```

```python
index = VectorStoreIndex.from_documents(docs)
```

Now **query engine** converts the index into a query engine, allowing you to run queries against the indexed documents.

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("can you give me the OGS contact information and the office location for Boston Campus?")
```

```python
Response = str(response)
print(response)
```

```python
formatted_response = '\n'.join(sentence.strip() for sentence in response.response.split('.') if sentence.strip())  # Access the response text via .response attribute
print(formatted_response)
```
